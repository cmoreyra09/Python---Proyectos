## Pivoteando & Unpivot

```python
import pandas as pd
from tabulate import tabulate

df = pd.read_csv(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv",delimiter=";")
df.head(5) ## Salida 5 Datos 
```

```python
## Agrupacion Aceite Ventas :

Filtro = df.loc[df['Categoria']== 'Aceites']
agrupacion = Filtro.groupby('Mes')['Ventas sin IGV'].sum()
agrupacion.to_frame().reset_index()
```

```python
## Ejercicio 1
## Se solicita realizar un cuadro resumen de las ventas por meses la cabeceras tienen que ser los meses y filas marca

pivoteando = df.pivot_table(index=['Categoria'],columns=['Mes'],values=['Ventas sin IGV'],aggfunc='sum').reset_index()
notnull = pivoteando.fillna('No se Encontro Registro') ## Cambiando los valores Nullos a un parametros String
notnull.columns = notnull.columns.droplevel() ## Para quitar las cabeceras que estan estan demas quedara una cabecera vacia.
notnull.rename(columns={'':'Categoria'},inplace=True) ## Cambiamos el nombre de campo vacio a un nombre real
notnull
```

```python
## Despivoteando el Ejercicio 1

# Cambiando el nombre de la columna 'Ventas Sin IGV' en el DataFrame pivoteado
pivoteando.columns = ['Categoria', 'Abril','Enero','Febrero','Juli','Junio','Marzo','Mayo']

# Despivotando los datos
despivoteado = pd.melt(pivoteando, id_vars=['Categoria'], value_name='Ventas Sin IGV', var_name='Mes')

# Mostrando el resultado
despivoteado
```

```python
## Ejercicio 2 : 

## Se solicita despivotear el siguiente dataframe

data = {
    'mes': ['Ene', 'Feb', 'Mar', 'Ene', 'Feb', 'Mar', 'Ene', 'Feb', 'Mar'],
    'categoria': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'ventas': [100, 200, 150, 75, 125, 100, 50, 75, 100]
}

df = pd.DataFrame(data)
```

```python
pivoteado = df.pivot_table(index=['categoria'], columns=['mes'], values=['ventas'], aggfunc='sum').reset_index()
pivoteado
```

```python
# Pivoteando los datos
pivoteado = df.pivot_table(index=['categoria'], columns=['mes'], values=['ventas'], aggfunc='sum').reset_index()

# Cambiando el nombre de la columna 'ventas' en el DataFrame pivoteado
pivoteado.columns = ['categoria', 'Ene', 'Feb', 'Mar']

# Despivotando los datos
despivoteado = pd.melt(pivoteado, id_vars=['categoria'], value_name='ventas', var_name='mes')

# Mostrando el resultado
despivoteado
```
