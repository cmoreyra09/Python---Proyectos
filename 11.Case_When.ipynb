```python
import pandas as pd
from tabulate import tabulate
import numpy as np

df = pd.read_csv(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv",delimiter=";")
df.head(5) ## Salida 5 Datos 
```

```python
## Ejercicio 1 :

## Se solicita agrupar las ventas por mes y obtener un resultado
cat_month = pd.CategoricalDtype(categories=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'], ordered=True)


agrupacion_v = df.groupby('Mes')['Ventas sin IGV'].count().reset_index()
## 9000 Correcto
## 8500 Promedio
## 8000 Bajo

def clasificacion_ventas(Ventas):
    if Ventas >= 90000:
        return Ventas*0
    elif Ventas >= 85000:
        return 'Promedio'
    else:
        return 'Bajo'

# Aplicar la función lambda usando apply y crear una nueva columna en el DataFrame
agrupacion_v['Ventas'] = agrupacion_v['Ventas sin IGV'].apply(lambda x: clasificacion_ventas(x))
agrupacion_v['Mes'] = agrupacion_v['Mes'].astype(cat_month)
final = agrupacion_v.sort_values(by='Mes',ascending=True)
final
```

```python
## Ejercicio 2 :
# Crear un DataFrame de ejemplo
# Crear un DataFrame de ejemplo
datos = {'nombre': ['Juan', 'María', 'Pedro', 'Ana', 'Luis'],
         'edad': [25, 35, 42, 19, 60],
         'genero': ['M', 'F', 'M', 'F', 'M']}
df = pd.DataFrame(datos)

# Utilizar .where() para crear una nueva columna "grupo_edad"
df['grupo_edad'] = df['edad'].where(df['edad'] < 30, 'Mayor de 30')

# Utilizar .where() para crear una nueva columna "grupo_edad_genero"
df['grupo_edad_genero'] = df['grupo_edad'].where((df['edad'] < 30) & (df['genero'] == 'M'), 'Mayor de 30 o mujer')

# Imprimir el DataFrame resultante
df
```
