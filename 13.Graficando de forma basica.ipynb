```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import numpy as np

df = pd.read_csv(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv",delimiter=";")
df.head(5) ## Salida 5 Datos 
```

```python
## Ejercicio 1:

## Se pide obtener las ventas de abrir ordenados por semana y mostrarlo en grafico de barras

Filtro = df.loc[df['Mes']== 'Abril']
agrupacion = Filtro.groupby('Semana')['Ventas sin IGV'].sum().reset_index()
Ordenamiento = agrupacion.sort_values(by='Semana',ascending=True)
Ordenamiento
```

```python
plt.bar(Ordenamiento['Semana'], Ordenamiento['Ventas sin IGV'], color='blue')
plt.xlabel('Semanas')
plt.ylabel('Ventas sin IGV')
plt.title('Ventas por semana')

# Agrega las etiquetas de datos con 'S/. '
for i, v in enumerate(Ordenamiento['Ventas sin IGV']):
    plt.text(i, v, 'S/. ' + str(v), ha='center')

# Agrega el signo de soles al eje Y
plt.ticklabel_format(style='plain', axis='y', useLocale=True)
plt.gca().yaxis.set_major_formatter('${x:,.0f}')

plt.show()
```
