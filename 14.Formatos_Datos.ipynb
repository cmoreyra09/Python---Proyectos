```python
## Paso 1 Conversiona a formato fecha datetime

df['Fechas'] = pd.to_datetime(df['Fechas'],format='%d/%m/%Y')
df.head(5)
```

```python
## Extracion de meses y dias de fechas en ingles

df['Mes_Nombre'] = df['Fechas'].dt.month_name() ## Meses
df['Dia_Nombre'] = df['Fechas'].dt.day_name() ## Dia
df.head(5)
```

```python
## Ahora se pide ver el tipo de formato que tiene los objetos correspondientes
df.info()
```

```python
## Mostrando la cantida de meses que tiene el dataset
x = df.groupby('Mes_Nombre')['Mes_Nombre'].count().reset_index(name='Cantidades_meses')
x
```

```python
## Para poder pasar los nombres de ingles a español utilizaremos una funcion que nos permita realizar la misma gestion :
## Funcion que me permite traducir los meses de ingles a español
## Funcion Meses

def meses_español(traduccion):
        meses = {
        "January": "Enero",
        "February": "Febrero",
        "March": "Marzo",
        "April": "Abril",
        "May": "Mayo",
        "June": "Junio",
        "July": "Julio",
        "August": "Agosto",
        "September": "Septiembre",
        "October": "Octubre",
        "November": "Noviembre",
        "December": "Diciembre"
        }
        if traduccion in meses:
                return meses [traduccion]
        else:
                return 'Mes No encontrado'
## Funcion dias
def dias_español(traduccion_d):
    dias  = {
        "Monday": "Lunes",
        "Tuesday": "Martes",
        "Wednesday": "Miercoles",
        "Thursday": "Jueves",
        "Friday": "Viernes",
        "Saturday": "Sabado",
        "Sunday": "Domingo"
    }
    if traduccion_d in dias:
        return dias[traduccion_d]
    else:
        return 'Dia No Encontrado'
```

```python
## Aplicamos la traduccion de los meses y dias en ingles a meses y dias en español
Meses = ['Mes_Nombre','Dia_Nombre']
df3 = df[Meses]
df3['Mes_Español'] = df3['Mes_Nombre'].apply(meses_español)
df3['Dia_Español'] = df3['Dia_Nombre'].apply(dias_español)
df_final = df3[['Mes_Español','Dia_Español']]
df_final
## El codigo se ejecuto de manera correcta:
## Significado del mensaje es : 
## Se produce cuando intentas asignar valores a una copia de un DataFrame en lugar de modificar el DataFrame original directamente. 
## Este aviso es una medida de precaución de Pandas para evitar posibles problemas de copias no deseadas.
```

```python
from tabulate import tabulate
cantidad_meses = df_final.groupby('Mes_Español').size().reset_index(name='Cantidad').sort_values('Cantidad',ascending=False) ## Se usa size cuando es solo una columna
cantidad_dias = df_final.groupby('Dia_Español').size().reset_index(name='Cantidad').sort_values(by='Cantidad',ascending=False) ## Se usa size cuando es solo una columna
total_mese = cantidad_meses['Cantidad'].sum() ## Se debe especificar la columna a sumar
total_dias = cantidad_dias['Cantidad'].sum() ## Se debe especificar la columna a sumar
meses_lista = cantidad_meses.values.tolist()
dias_lista  = cantidad_dias.values.tolist()

print('------------------------s---------------------')
print('***************- Resumen Meses -**************')
print('----------------------------------------------')
print('Cantidad_Meses')
print(f'La suma de las cantidad es de {total_mese}   ')
print('----------------------------------------------')
print(tabulate(meses_lista,headers=cantidad_meses,tablefmt='psql'))
print()
print('----------------------------------------------')
print('Cantidad_Dias')
print(f'La suma de las cantidad es de {total_dias} dias')
print('----------------------------------------------')
print(tabulate(dias_lista,headers=cantidad_dias.columns,tablefmt='psql'))
```
