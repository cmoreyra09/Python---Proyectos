## Identificacion de Duplicados , Eliminacion -> Pre ETL

```python
import pandas as pd
from tabulate import tabulate

df = pd.read_excel(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Identificar Duplicaods.xlsx")
df.head(5)
```

```python
## Cantidad de registros totales
print(f'La cantidad del dataset es de {df.shape[0]} registros') ## El siguiente parametro realiza la cuenta considerando la primera columna en este caso seria fechas
```

```python
##  Ejercicio 1 : 
    ## Se solicita obtener la data final sin duplicados. Considerar que en total deberia se deberia obtener 70638 Registros Mostrarlo en formato datafram y tabulado.
    ## Asi mismo, se pide que se cree el output final donde el nombre del archivo sera Data_No_Duplicada el archivo de salida debe ser Plano (csv)
    ## Realizarlo de las 2 formas posibles
    
## Forma 1

ejercicio_1 = df.drop_duplicates(subset=df.columns) ## En este algoritmo se esta considerando todos las columnas de manera mas automatica
ejercicio_1
    
```

```python
## Forma 2

eje1_f2 = df.drop_duplicates(subset=['Fechas','Codigo_Producto','Categoria_Producto','Nombre_Producto','Marca_Producto'])
eje1_f2
```

```python
## Exportando el archivo ouput sin duplicidades de manera profesional como ingeniero de software
## Creando las variables para la salida correspondiente

despositorio_ruta = r'C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada'
nombre_archivo = 'Data_No_Duplicada.csv'

for i in range (1,2):
    try:
        if i == 1:
            eje1_f2.to_csv(r'C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Data_No_Duplicada.csv',index=False)  ## Index = false para evitar los indices de cada registro
            print(f'El archivo {nombre_archivo} se guardo exitosamente en la ruta {despositorio_ruta}')
    except Exception as e:
        print(f'Existe error al guardar el archivo,favor de revisar.')
```

```python
## Ejercicio 2 :
    ## Se solicita obtener unicamente los registros que se estan duplicando y colocarlo en un archivo csv en la ruta de BBDD - No Estructurada

Ejercicio_2 = df[df.duplicated(keep='last')]
Ejercicio_2 ## Resultado
```

```python
## Subscribiendo los mismo datos al mismo data set 
consolidado = pd.concat([ejercicio_1,Ejercicio_2])
consolidado
```

```python
## Exportando los datos duplicados aislados a la ruta correspondiente :

ruta = r'C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada'
output = 'Data_Duplicada_Aislada.csv'

for i in range (1,2):
    try:
        if i ==1:
            Ejercicio_2.to_csv(r'C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Data_Duplicada_Aislada.csv',index=False)
            print(f'El archivo {output} se guardo de manera exitosa en la ruta {ruta}')
    except Exception as e:
        print('Surgio un problema en el guardado del nuevo dataset, favor de revisar el codigo correspondientemente')
```
