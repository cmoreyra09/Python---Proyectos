## Ordenando columnas y tops

```python
import pandas as pd
from tabulate import tabulate

df = pd.read_csv(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv",delimiter=";")
df.head(5) ## Salida 5 Datos 
```

```python
##  Ejercicio 1
    # Forma Pandas
    # Se solcita tener un top 5 de los registros con mayor ventas teniendo en cuentas ambas librerias.

df_ejercicio1 = df.sort_values(by='Ventas sin IGV',ascending= False).head(5)
df_ejercicio1
```

```python
## Ejercicio 2
    # Forma tabulate

atributos = ['Año', 'Mes', 'Semana', 'Fechas', 'Cadena', 'Departamento', 'Distrito',
       'Categoria', 'Marca', 'Codigo_Producto', 'Material', 'Unidades B2B',
       'Peso en Toneladas', 'Ventas sin IGV']

df2 = df[atributos].sort_values(by='Ventas sin IGV',ascending= False)
df2_1 = df2.values.tolist()
top_5 = df2_1[:5]
print(tabulate(top_5,headers=df2,tablefmt='psql'))
```
