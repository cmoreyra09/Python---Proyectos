## Realizando Filtros en Python

```python
import pandas as pd
from tabulate import tabulate

def read_filepath(file):
    data = pd.read_csv(file,delimiter=';')
    return data

def tabulacion(data):
    x = print(tabulate(data,headers='keys',tablefmt='rounded_outline'))
    return x 

df = read_filepath(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv")
tabulacion(df.head(5))
```

```python
df[df['Semana']== 'Semana 1'].count()
```

```python
## Ejercicio 1 

## Se solicita obtener los registros los registros que tengan 100 unidades vendidas del mes Enero una ves obtenido el resultado .
## Ademas se solicita mostrar un top 10 de las de los registros encontrados y
## Ordenarlo de mayor a menor segun las ventas.Ademnas se solicita despleguarlo por las dos librerias

eje1 = df.loc[(df['Mes'] == 'Enero') & (df['Unidades B2B']==100)].sort_values(by='Ventas sin IGV',ascending=False).head(10)
tabulacion(eje1)
```

```python
## Forma 2 Tabulate

Atributos_1 = ['Año', 'Mes', 'Semana', 'Fechas', 'Cadena', 'Departamento', 'Distrito','Categoria', 'Marca', 'Codigo_Producto', 'Material', 'Unidades B2B',
       'Peso en Toneladas', 'Ventas sin IGV']

df_2 = df[Atributos_1].sort_values(by= 'Ventas sin IGV',ascending=False)
df_3 = df_2.loc[(df_2['Unidades B2B'] == 100)].values.tolist()
df_filtro = df_3[:10]

print(tabulate(df_filtro,headers=Atributos_1,tablefmt='psql'))

```

```python
## Ejercicio 2 

## Se solicita obtener las ventas ordenadas de mayor a menor unicamente de la semanan 1 un top 10 Realizarlo en ambas formas con tabulate

eje2 = df.loc[(df['Semana']== 'Semana 1')&(df['Ventas sin IGV'] > df['Ventas sin IGV'].mean())].sort_values(by='Ventas sin IGV',ascending=False).head(10)
tabulacion(eje2)


'''
Si se desea ordenar un atributo en python existe diferencia entre un valor numerico y un valor string(Caracteres)

En situacion de String se usa lo siguiente :

- Filtro_1.sort_values(by='Semana', ascending=True)

  En situacion numerica se usa lo siguiente :

- Filtro_2.sort_values(by='Ventas sin IGV',ascending=False) 

'''
```

```python
## Forma 2
    # Tabulate

filtro_f23 = df[Atributos_1].sort_values(by='Ventas sin IGV',ascending=False) ## Ordernando de mayor a menor mediante el campo Ventas sin IGV
filtro_f24 = filtro_f23.loc[filtro_f23['Semana']== 'Semana 1'].values.tolist() ## Filtro y conversion a lista
filtro_25 = filtro_f24[:10]  ## Top 10
print(tabulate(filtro_25,headers=filtro_f23,tablefmt='psql')) ## Mostrando Resultado
```

```python
## Otra forma de filtrado usando df.query

Filtro_2_ordenando_ascendente_1 = df.query("Marca == 'Primor' and Distrito == 'Huancayo' and Codigo_Producto == 3201006 ").sort_values(by='Ventas sin IGV',ascending=False).head(10)
Filtro_2_ordenando_ascendente_1
```

```python
## Ejercicio 3 :

## Se solicita ver los registros unicamento de la categoria Lejias, departamento lima, y en relacion a las unidades vendidas se solicita tener mostrar mayor a 100 cantidades
## Mostrar unicamente los 10 primeros registros que tengan mayor ventas

df_lejias = df.loc[(df['Categoria'] == 'Lejias') & (df['Departamento']== 'Lima') & (df['Unidades B2B'] > 100)].sort_values(by='Ventas sin IGV',ascending=False).head(10)
df_lejias
```

```python
## Mostrar todos los dataframe con tabulate y solo mostrar los 10 primeros registros que contenga una mayor cantidad de ventas
data = ['Año','Mes','Semana','Fechas','Cadena','Departamento','Distrito','Categoria','Marca','Codigo_Producto','Material','Unidades B2B','Peso en Toneladas','Ventas sin IGV']
data_2 = df[data].sort_values(by='Ventas sin IGV',ascending=False).head(10).values.tolist()
print(tabulate(data_2,headers=data,tablefmt='psql'))
print(tabulate)
```

```python
## Filtros como where in and not in en SQL-server (IN)

## Ejercicio 4

## Se solicita obtener unicamento las ventas de lima y lambayque de la primera semana 

filtro_ejercicio_4 = df.loc[(df['Semana'] == 'Semana 1') & (df['Departamento'].isin(['Lima','Lambayeque']))].sort_values(by='Ventas sin IGV',ascending=False).head(10)
filtro_ejercicio_4
```

```python
## Forma tabulando

tabulado_1 = filtro_ejercicio_4.values.tolist()
print(tabulate(tabulado_1,headers=df.columns,tablefmt='psql'))
```

```python
## Filtros como where in and not in en SQL-server (NOT IN)

## Ejercicio 5

## Se solicita obtener unicamento las ventas que no tengan la ciudad de lima, lambayque y piura de la primera semana 

filtro_ejercicio_5 = df.loc[(df['Semana']== 'Semana 1') & (-df['Departamento'].isin(['Lima','Lambayeque','Piura']))].sort_values(by='Ventas sin IGV',ascending=False).head(5)
filtro_ejercicio_5

```

```python
## Forma Tabulando

tabulado_2 = filtro_ejercicio_5.values.tolist()
print(tabulate(tabulado_2,headers=df.columns,tablefmt='psql'))
```
