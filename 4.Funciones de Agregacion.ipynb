```python
import pandas as pd
from tabulate import tabulate

df = pd.read_csv(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv",delimiter=";")
df.head(5) ## Salida 5 Datos 
```

```python
## Ejercicio 1

## En esta seccion realizaremos el uso de funciones de agreacion de acuerdo al dataset Makro_Ventas

## Se solicita obtener lo siguiente : 

## Maximo , Minimo , Promedio , moda , mediana

maximo = df['Ventas sin IGV'].max()
idmaximo = df['Ventas sin IGV'].idxmax()

minimo = df['Ventas sin IGV'].min()
idminimo = df['Ventas sin IGV'].idxmin()

promedio = df['Ventas sin IGV'].mean()

mediana_1 = df.sort_values(by='Ventas sin IGV',ascending=False)
mediana_2 = mediana_1['Ventas sin IGV'].median()

moda = df['Ventas sin IGV'].mode()
moda
print('---------')
print('RESULTADO')
print('---------')
print(f' - El valor maximo de ventas es {maximo} y su id correspondiente es {idmaximo}')
print(f' - El valor maximo de ventas es {minimo} y su id correspondiente es {idminimo}')
print(f' - El promedio de ventas es de {promedio.round(2)}')
print(f' - La mediana correspondiente de las ventas es {mediana_2.round(2)}')
print(f' - La moda correspondiente de las ventas es {moda}')
```
