## Funciones de Agregacion

```python
## Funciones de Agregacion Group BY
import matplotlib.pyplot as plt
import pandas as pd
from tabulate import tabulate

df = pd.read_csv(r"C:\Users\User\OneDrive\Documentos\Archivos_Carlos\BBDD - No estructurada\Makro_Ventas.csv",delimiter=";")
df.head(5) ## Salida 5 Datos 

##df = pd.read_csv('Makro_Ventas.csv',delimiter = ";")
##df.head(5)
```

```python
## Ejercicio 1 :

## Se solicita obtener la cantidad de semanas que participan en este data set
## Mostrar cona las dos librerias

ejercicio_1 = df.groupby('Semana')['Semana'].count().reset_index(name='Cantidad de Semanas')
ordenamiento = ejercicio_1.sort_values(by='Cantidad de Semanas',ascending=True)
ordenamiento
```

```python
## Forma 1 -- Tabulate con lista
lista = ordenamiento.values.tolist()
filtro = lista[:2]
print(tabulate(filtro,headers=ejercicio_1.columns,tablefmt='psql'))
```

```python
## Forma 2 -- Tabulate sin lista forma data frame
print(tabulate(ordenamiento,headers=ejercicio_1.columns,tablefmt='psql'))
```

```python
## Ejercicio 2:

## Se solicita obtener las ventas totales por todos los meses del data set, asi mismo se debe de ordenar de mayor a menor y mostrar de las 2 formas correspondiente
cat_month = pd.CategoricalDtype(categories=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio'], ordered=True)

ejercicio_2 = df.groupby('Mes')['Ventas sin IGV'].sum().reset_index(name='Ventas por Mes')
ejercicio_2.sort_values(by=['Mes','Ventas por Mes'],ascending = [True,True]).sort_values(by='Mes',ascending=True)
ejercicio_2['Mes'] = ejercicio_2['Mes'].astype(cat_month)
ejercicio_2.sort_values(by='Mes',ascending=True)

```

```python
## Ejercicio 3 

## Se solicita mostrar la ventas maximas por meses,semanas y categorias mostrarlo en los 2 formatos
## Considerar los filtros de semana 1 y 2 
cat_month = pd.CategoricalDtype(categories=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio'], ordered=True)

Eje_3 = df.groupby(['Mes','Semana','Categoria'])['Ventas sin IGV'].max().reset_index(name='Ventas por mes')
Filtro_1 = Eje_3.loc[Eje_3['Semana'].isin(['Semana 1','Semana 2'])].sort_values(by='Ventas por mes',ascending=False)
Filtro_1['Mes'] = Filtro_1['Mes'].astype(cat_month) ## Traspaso para ordenar de manera ordenada los meses
final = Filtro_1.sort_values(by=['Mes','Ventas por mes'],ascending=[True,False])
final
```

```python
## Forma 2 : Tabulate

tabulado_1 = final.values.tolist()
print(tabulate(tabulado_1,headers=final.columns,tablefmt='psql'))
```

```python
## Ejercicio 4
## Se solicita obtener las suma de ventas por mes semana y categoria y mostrarlo tabulado y normal

ejercicio_3 = df.groupby(['Mes','Semana','Categoria'])['Ventas sin IGV'].sum().reset_index(name='Ventas')
lista_eje3 = ejercicio_3.values.tolist()
print(tabulate(lista_eje3,headers=ejercicio_3.columns,tablefmt='psql'))
```

```python
## Forma Normal 1 

forma_normal = df.groupby(['Mes','Semana','Categoria'])['Ventas sin IGV'].sum()
forma_normal.to_frame()
```

```python
## Forma Normal 2

forma_normal = df.groupby(['Mes','Semana','Categoria'])['Ventas sin IGV'].sum().reset_index(name='Ventas Sin IGV 2')
forma_normal
```

```python
# Definir el orden de las categorías del mes
cat_month = pd.CategoricalDtype(categories=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio'], ordered=True)

# Agrupar y sumar las ventas por mes
ejercicio_2 = df.groupby('Mes')['Ventas sin IGV'].sum().reset_index(name='Ventas por Mes')

# Convertir la columna "Mes" en una variable categórica y ordenarla
ejercicio_2['Mes'] = ejercicio_2['Mes'].astype(cat_month)
ejercicio_2 = ejercicio_2.sort_values(by='Mes')

# Mostrar el DataFrame ordenado
ejercicio_2
```

```python
## Ejercicio 5 

## Se desea obtener la cantidad de semanas que se estan registrando en el dataset correspondiente

x = df.groupby('Semana')['Semana'].count()
x.to_frame().rename(columns={'Semana':'Cantidad de Semanas'})
```

```python
## Ejercicio 6

## Se pide las ventas por semana  con id
ventas = df.groupby('Semana')['Ventas sin IGV'].sum().reset_index(name='Suma de Ventas por Semana')
ventas['Suma de Ventas por Semana'] = ventas['Suma de Ventas por Semana'].astype(int)
print(tabulate(ventas, headers=ventas.columns, tablefmt='psql', floatfmt='d'))
```

```python
## Ejercicio 7 -- Tabla Resumen
## Se solicita realizar una agrupacion en relacion a las ventas por categorias 
## Es decir se debe de presentar las ventas por categorias teniendo en cuenta, valores maximo,valaores minimo, promedio y recuendo mostrar en formato tabulate y dataframe

ejercicio_7 = df.groupby('Categoria')['Ventas sin IGV'].agg(['sum','max','min','mean','count'])
ejercicio_7_2 = ejercicio_7.rename(columns={'sum':'Suma Ventas','max':'Maxima Ventas','min':'Minima Ventas','mean':'Ventas Promedio','count':'Recuento Categoria'}).reset_index()
ejercicio_7_2_1 = ejercicio_7_2.sort_values(by='Suma Ventas',ascending=False)
ejercicio_7_2_1
```

```python
## Mostrando Resultado en formato tabulate
ejercicio_7_2_1['Suma Ventas'] = ejercicio_7_2_1['Suma Ventas'].astype(int) 
tabulacion_7 = ejercicio_7_2_1.values.tolist()
print(tabulate(tabulacion_7,headers=ejercicio_7_2_1.columns,tablefmt='psql'))
```
