```python
import pandas as pd
from tabulate import tabulate

## En esta seccion realizaremos una simulacion con 2 supuestos dataset donde realizaremos la teorioa de joins

## DataFrame Ejemplos

# Crear dos DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df2 = pd.DataFrame({'A': ['A0', 'A1', 'A6', 'A7'],
                    'F': ['B4', 'B5', 'B6', 'B7'],
                    'G': ['C4', 'C5', 'C6', 'C7'],
                    'H': ['D4', 'D5', 'D6', 'D7']})

df1
```

```python
df1
```

```python
df2
```

```python
## Ejemplo 1 : Sintaxis

# Realizar left join por columna "A"
left_join = pd.merge(df1, df2, on='A', how='left')
left_join
```

```python
## Ejemplo 2 : Sintaxis

# Realizar inner join por columna "A"
inner_join = pd.merge(df1, df2, on='A', how='inner')
inner_join
```

```python
## Realizando una operacion de filtro en relacion al resultado obtenido

x = inner_join.loc[inner_join['A']== 'A1'].sort_values(by='A',ascending=False)
x[['A','D']]
```

```python
## Ejercicio 1 : 

## Mostrar los duplicados de los atributos de codigo producto y material y proceder a realizar la normalizacion.

ejercicio_1 = df[['Codigo_Producto','Material']]
x_1 = ejercicio_1[ejercicio_1.duplicated(['Codigo_Producto','Material'],keep='last')].sort_values(by='Codigo_Producto',ascending=False)
x_1
```

```python
## Mostrando los con formato tabulate

x_2 = x_1.values.tolist()
filtro = x_2[:5]
print(tabulate(filtro,headers=x_1.columns,tablefmt='psql'))
```

```python
## Ejercicio 2
## Si pide crear una tabla donde se obtenga el primary key como codigo y los materiales se de llamar detalle.Una vez creada la tabla, se tiene que crear otra tabla donde se llame ventas y solo esten el codigo 
## de producto y variables numericas

Productos_ = df[['Codigo_Producto','Material']].drop_duplicates() ## Se quita la parte de duplicados para proceder con la normalizacion
Productos_[['Codigo_Producto','Material']].count() ## Se realiza el conteo estadistico para confirma ambas cantidades esten correcta
```

```python
## Creacion de la tabla Productos con el nuevo nombre del atributo Detalle

Productos_ = df[['Codigo_Producto','Material']].drop_duplicates()
## Renombrando la columna Material a Detalle
cambio_nombre = Productos_.rename(columns={'Material':'Detalle'})
eliminacion_duplicados = cambio_nombre.drop_duplicates(['Codigo_Producto'])
Producto = eliminacion_duplicados
Producto
```

```python
## Creacion de la tabla Codigo_Producto mas valores estadisticos

ventas_ = df[['Codigo_Producto','Unidades B2B','Peso en Toneladas','Ventas sin IGV']]
ventas = ventas_.rename(columns={'Codigo_Producto':'Codigo'})
ventas
```

```python
## Ultimo Paso
## Se precisa encontrar los materiales de lo que se vendieron 

consolidado = pd.merge(ventas,Producto,left_on='Codigo',right_on='Codigo_Producto',how='inner')
consolidado_final = consolidado[['Codigo','Detalle','Unidades B2B', 'Peso en Toneladas', 'Ventas sin IGV']]
final_consolidado = consolidado_final.values.tolist()
final_consolidado = final_consolidado[:5]
print(tabulate(final_consolidado,headers=consolidado_final.columns,tablefmt='psql'))
```
