## Proyecto de Analisis de Concentración – Tablas de Frecuencias para identificar el desempeño de diferentes variables 


```python
from tabulate import tabulate
import pandas as pd
import numpy as np

## Importacion del datase

df = pd.read_csv('Concursos.csv',delimiter=';')
```

```python
## Dimensiones
df.info()
```

```python
## Variables de Concentracion

columnas = df.columns.tolist()
tipos = df.dtypes.tolist()

tabla = {'Variables': columnas, 'Tipos': tipos}
print(tabulate(tabla, headers='keys', tablefmt='psql'))
```

```python
# Obtener la frecuencia relativa y absoluta de cada columna y mostrar en una tabla
print('')
print('---------------------------')
print('Analisis de Concentracion:')
print('----------------------------')
print('')

for col in df.columns:
    freq_abs = df[col].value_counts().sort_index()
    freq_rel = df[col].value_counts(normalize=True).sort_index() * 100
    freq_rel = freq_rel.round(2).apply(lambda x: f"{x}%")
    freq_table = pd.concat([freq_abs, freq_rel], axis=1, keys=["Frecuencia Absoluta", "Frecuencia Relativa"])
    freq_table = freq_table.sort_values("Frecuencia Relativa", ascending=False).sort_values("Frecuencia Absoluta", ascending=False)
    print(f"Tabla de frecuencias de {col}:")
    print(tabulate(freq_table, headers='keys', tablefmt='psql'))
    print()
```
